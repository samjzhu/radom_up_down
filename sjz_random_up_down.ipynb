```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def randomUpDown(k):
    print(f'Samples={k}')
    arr = np.array(np.random.randint(2, size=k))
    # print(arr)
    c = np.count_nonzero(arr)
    print(f'Count of Ones: {c}')

    # Replace all 0 with -1
    v_arr = np.where(arr == 0, -1, arr)
    # print(v_arr)

    # Sum of values
    s_arr = []
    s = 0
    # continuous distribution
    dist = {}
    last_v = 0
    continuous_count = 0

    for v in v_arr:
        s += v
        s_arr.append(s)
        if v == last_v:
            continuous_count += 1
        else:
            current_sum = dist.get(continuous_count)
            if current_sum is None:
                current_sum = 0
            dist[continuous_count] = current_sum + 1
            continuous_count = 0
            last_v = v
    return s_arr, sorted(dist.items())
```

```python
k = 100000
```

```python
v, d = randomUpDown(1000)
```

```python
print(f'Final value:{v[-1]}')
plt.plot(v)
plt.show()
```

```python
print(f'continuous distribution:{d}')
plt.plot(np.array(d)[:, 0], np.array(d)[:, 1])
plt.show()
```
