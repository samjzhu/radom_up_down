```python
k = 100000
print(f'Samples={k}')
arr = np.array(np.random.randint(2, size=k))
# print(arr)
```

```python
c = np.count_nonzero(arr)
print(f'Count of Ones: {c}')
```

```python
# Replace all 0 with -1
v_arr = np.where(arr == 0, -1, arr)
# print(v_arr)
```

```python
# Sum of values
s_arr = []
s = 0

# continuous distribution
dist = {}
last_v = 0
continuous_count = 0

for v in v_arr:
    #fill sum of values series
    s += v
    s_arr.append(s)
    
    #collect continuous_count distribution
    if v == last_v:
        continuous_count += 1
    else:
        current_sum = dist.get(continuous_count)
        if current_sum is None:
            current_sum = 0
        dist[continuous_count] = current_sum + 1
        continuous_count = 0
        last_v = v


#print(s_arr)
```

```python
#sum of all values
sum(v_arr)
```

```python
#The last value should be same as the sum
s_arr[-1]
```

```python
plt.plot(s_arr)
plt.show()
```
